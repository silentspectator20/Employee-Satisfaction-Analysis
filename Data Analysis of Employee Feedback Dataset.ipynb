## Data Preprocessing and Feature Engineering

Since there are too many columns interrelated so we can combine them as follows:

## Statistical Tests

```python
import pandas as pd
from scipy.stats import f_oneway, ttest_ind

# List of questions/columns to test
questions = [
       'I_receive_clear_instructions_and_guidance_from_my_supervisors',
       'I_feel_comfortable_approaching_my_supervisors_with_questions_or_concerns',
       'The_organization_values_open_and_honest_communication',
       'The_organization_encourages_a_culture_of_providing_rewards_for_good_work',
       'My_Ideas_and_opinions_are_valued_by_my_manager',
       'Safety_training_is_conducted_on_a_regular_basis_for_all_employees',
       'I_get_the_opportunity_to_participate_in_a_structured_training_program_conducted_in_the_Organization',
       'I_get_on_the_job_training_by_my_supervisor',
       'I_try_to_acquire_new_knowledge_and_new_skills_through_self_learning_and_peer_learning',
       'This_is_a_physically_safe_place_to_work',
       'While_on_work_utmost_priority_is_given_on_the_safety',
       'I_feel_that_there_is_no_discrimination_and_harassment_in_the_organization',
       'I_am_satisfied_with_health_insurance_and_medical_insurance_provided_by_the_organization',
       'I_am_satisfied_with_company’s_leave_policy_(sick_leave,_paid_leave_etc)',
       'I_get_my_salary/wages_on_time',
       'The_compensation,_benefits_and_facilities_provided_by_the_organization_is_better_than_among_the_top_three_in_Asansol'
]

# Perform ANOVA test for each question to check if there's any significant difference based on zones
zones = ['Zone 1', 'Zone 2', 'Zone 3', 'Zone 4']
for question in questions:
    anova_data = [df[df['Zone'] == zone][question] for zone in zones]
    anova_result = f_oneway(*anova_data)
    print(f"\nQuestion: {question}")
    print("ANOVA Test:")
    print("Null hypothesis (H0): Responses of employees from all 4 zones are the same.")
    print("Alternative hypothesis (H1): Responses of employees from all 4 zones are not the same.")
    print("p-value:", anova_result.pvalue)
    if anova_result.pvalue < 0.05:
        print("Result: Reject null hypothesis, there is a significant difference in responses among zones.")
    else:
        print("Result: Fail to reject null hypothesis, responses are not significantly different among zones.")

    # Perform t-tests between pairs of zones for each question
    for i in range(len(zones)):
        for j in range(i + 1, len(zones)):
            zone_i = zones[i]
            zone_j = zones[j]
            ttest_result = ttest_ind(
                df[df['Zone'] == zone_i][question],
                df[df['Zone'] == zone_j][question]
            )
            print(f"\nT-Test between {zone_i} and {zone_j}:")
            print("Null hypothesis (H0): Responses of employees from", zone_i, "and", zone_j, "are the same.")
            print("Alternative hypothesis (H1): Responses of employees from", zone_i, "and", zone_j, "are not the same.")
            print("p-value:", ttest_result.pvalue)
            if ttest_result.pvalue < 0.05:
                print("Result: Reject null hypothesis, there is a significant difference in responses.")
            else:
                print("Result: Fail to reject null hypothesis, responses are not significantly different.")
```
